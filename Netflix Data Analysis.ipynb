### 2. In which year highest number of Tv shows and  were released? Show with bar graph.

```python
df['Release_Date']=pd.to_datetime(df['Release_Date'])
```

```python
df.dtypes
```

```python
df['year']=df['Release_Date'].dt.year
```

```python
df
```

```python
top_year=df.groupby(['year'],as_index=False)['Category'].count().sort_values(by='Category', ascending=False)
top_year
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='year',y='Category',data=top_year)
plt.show()
```

### 4. Display all movies that were released in the year 2020

### 5.Show titles of TV shows released in India only

```python
df[  (df['Country']=='India') & (df['Category']=='TV Show') ]['Title'].reset_index()
```

### 6.Top 10 directors by movies

```python
df[df['Category']=='Movie'].groupby(['Director'],as_index=False)['Category'].count().reset_index().sort_values(by='Category',ascending=False)[0:10]
```

### 7.Top 10 directors by TV Shows

```python
df[df['Category']=='TV Show'].groupby(['Director'],as_index=False)['Category'].count().reset_index().sort_values(by='Category',ascending=False)[0:10]
```

### 8.Top 10 directors by both movies and TV shows

```python
df.groupby(['Director'],as_index=False)['Category'].count().reset_index().sort_values(by='Category',ascending=False)[0:10]
```
